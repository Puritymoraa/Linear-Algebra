```python
import numpy as np
```

```python
A = np.array([
    [0.65, 0.10, 0.25, 0.00],
    [0.30, 0.10, 0.35, 0.25],
    [0.30, 0.15, 0.15, 0.40],
    [0.20, 0.10, 0.30, 0.40]
])
```

```python
I = np.eye(4)
IA_Difference = I - A
```

```python
b = np.zeros(4)
# Use numpy's linear algebra solver to find equilibrium prices
prices = np.linalg.solve(IA_Difference, b)
labels = ["Agriculture", "Energy", "Manufacturing", "Transportation"]
for i in range(4):
    price = prices[i]
    print(f"{labels[i]}: {price}")
```

```python
# Plotting multiple lines using sin and cos
x = np.linspace(-np.pi, np.pi,100)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
# We can as well simply combine the multiple lines in one line of code
plt.plot(x, np.sin(x), x, np.cos(x))
plt.show()
```

```python
#adding labels to the curves
plt.plot(x, np.sin(x), label='sine')
plt.plot(x, np.cos(x), label='cosine')
plt.legend()
plt.title('Sine and Cosine functions')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Plotting vectors

```python
#Ploting vectors using quiver: Ploting the vectors [1,2] and [4,3]
plt.quiver(0,0,1,2, color='r', angles = 'xy', scale_units= 'xy', scale =1)
plt.quiver(0,0,4,3, color = 'b', angles = 'xy', scale_units= 'xy', scale =1)
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.plot([-2,2], [0,0], '-k')
plt.plot([0,0], [-2,2], 'k')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
u = np.array([1,0.5])
v = 3*u
plt.quiver(0,0,u[0],u[1], color='r', angles = 'xy', scale_units= 'xy', scale =1)
plt.quiver(0,0,v[0],v[1], color = 'b', angles = 'xy', scale_units= 'xy', scale =1)

plt.text(3,2, 'v = 3u', color = 'b') #adding text to the graph
#Displaying the origin as a black dot
plt.plot(0,0, 'ko')
plt.plot([-6,6], [0,0], 'm--', [0,0],[-5,5], 'm--')
plt.xlabel('x')
plt.ylabel('y')

for c in range(-6,5):
  v = c*u
  plt.plot(v[0], v[1], 'b.')
plt.xlim(-6,6)
plt.ylim(-5,5)
plt.show()
```

Linear combinations of u and v:
cu+dv E Span{u,v}

```python
  fig = plt.figure()
u = np.array([1,1,2])
v= np.array([2, 1, -1])
ax = fig.add_subplot(projection = '3d')
plt.quiver(0,0,0,u[0],u[1],u[2], color='r')
plt.quiver(0,0,0,v[0],v[1],v[2], color = 'b')

#add the origin
ax.plot([0],[0],[0], 'ro')

for c in np.linspace(-2,2,20):
  for d in np.linspace(-2,2,20):
    w =c*u + d*v
    ax.plot3D(w[0],w[1],w[2], 'g.')
ax.set_xlim(-5,5)
ax.set_ylim(-5,5)
ax.set_zlim(-5,5)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```
